```python
import pandas as pd
```

```python
bike_sharing = pd.read_csv("C:/bike-sharing-project/seoul-bike-sharing-data.csv", encoding='cp949') # Korean encoding
bike_sharing.head()
```

```python
# Create DateTime column, drop Date and Hour columns
bike_sharing['DateTime'] = pd.to_datetime(
    bike_sharing['Date'] + ' ' + bike_sharing['Hour'].astype(str),
    format='%d/%m/%Y %H'
)
bike_sharing.drop(columns=['Date', 'Hour'], inplace=True)
bike_sharing.set_index('DateTime', inplace=True)
```

```python
# Rename columns for easier access
bike_sharing.rename(columns={
    'Rented Bike Count': 'rented_bike_ct',
    'Temperature(캜)': 'temp',
    'Humidity(%)': 'humidity', 
    'Wind speed (m/s)': 'wind_speed',
    'Visibility (10m)': 'visibility',
    'Dew point temperature(캜)': 'dew_point_temp',
    'Solar Radiation (MJ/m2)': 'solar_radiation',
    'Rainfall(mm)': 'rainfall',
    'Snowfall (cm)': 'snowfall',
    'Seasons': 'season',
    'Holiday': 'holiday',
    'Functioning Day': 'funct_day'
}, inplace=True)
```

```python
bike_sharing.info()
# Change object types to strings 
bike_sharing['season'] = bike_sharing['season'].astype(str)
bike_sharing['holiday'] = bike_sharing['holiday'].astype(str)
bike_sharing['funct_day'] = bike_sharing['funct_day'].astype(str)

# There's no missing data, and all columns are numeric except for 'season', 'holiday', and 'funct_day'
```
